```python
import os
from pprint import pprint

from dotenv import load_dotenv
from rich import inspect, pretty
from rich.console import Console

# set env vars
load_dotenv("./.env")
os.environ["LANGSMITH_PROJECT"] = "langchain_chatbot"

# setup pretty printing
console = Console()
pretty.install()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="gpt-5", model_provider="openai")
```

```python
# simple run
from langchain_core.messages import HumanMessage

_response = model.invoke([HumanMessage("What's your knowledge cutoff date?")])

pprint(_response)
```

```python
console.print(
    _response,
)
```

```python
inspect(_response, methods=True)
```
